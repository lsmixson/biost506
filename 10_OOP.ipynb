## Object oriented programming

- So far, we have mainly used the functionality provided by Python and its packages.
- Much of this is organized through objects, which come with convenient methods and attributes.
- What we have done so far works well, but to interact more deeply with these packages (and eventually write our own code in a similar style) we need to understand these ideas a bit better.

Object-oriented programming organizes code around objects:
- Class: A class is a blueprint that defines what something is and what it can do.
- Object: An object is a specific instance of a class.
- Attribute: An attribute is a piece of data stored in an object, such as a name or a value.
- Method: A method is a function associated with an object that defines something the object can do.

One simple example is a list. A list is an **object** created from Python’s ``list`` class. It has methods, such as ``.append()`` and ``.sort()``, which allow us to modify and work with the data it contains.

```python
subjects = ['subject_000', 'subject_007', 'subject_006', 'subject_002']
print(subjects)
subjects.sort() # this is a method that changes subjects
print(subjects)
subjects.reverse()
print(subjects)
subjects.pop()
print(subjects)
subjects.append('subject_010')
print(subjects)
```

# Making our own objects with methods

- Let us illustrate these ideas with a simple example: a class for **complex numbers**
- Unlike a real number, a complex number has two parts: a **real part** and an **imaginary part**
- A number such as $1 + 2i$ can therefore be represented by the pair $(1, 2)$, together with methods that define how complex numbers behave, such as addition or multiplication

It turns out that Python already has such a class, but we will build our own to illustrate the process.

```python
# Python native class
x = 10 + 5j
print(x.real)
print(x.imag)
print(x.conjugate()) # Operation that reverses the sign of the imaginary part
```

Let’s create our own version of a complex number.

```python
class mycomplex:
    def __init__(self, real=0, imag=0):
        self.r = real
        self.i = imag

    def conjugate(self): #note this modifies self
        self.i =  -self.i

    def print(self):
        print((self.r, self.i))

y = mycomplex(10,5)
y.print()
y.conjugate()
y.print()
```

```python
y = mycomplex()
y.print()
```

Let's now create a version that doesn't modify the object when we conjugate that allows you to choose whether to make changes inplace or return conjugated complex number

```python
class mycomplex:
    def __init__(self, real, imag):
        self.r = real
        self.i = imag

    # def conjugate(self, inplace = False): 
    #    To be completed
        
    def print(self):
        print((self.r, self.i))

y = mycomplex(10,5)
y.print()
z = y.conjugate()
y.print()
z.print()
```

# Special methods, for example `__str__`

Python classes can define **special methods** to control how objects behave in built-in operations. These methods have names that begin and end with double underscores.

For example:

- `__str__` defines what is shown when we print an object
- `__add__` defines what happens when we use `+`
- `__len__` defines what happens when we call `len(...)`

A useful example is `__str__`: if we define it, then `print(obj)` can display a clear and readable description of the object instead of the default output.
