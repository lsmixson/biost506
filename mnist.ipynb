## MNIST

MNIST is a dataset of handwritten digits that is commonly used for training and testing machine learning models. It consists of 60,000 training images and 10,000 test images, each of which is a 28x28 pixel grayscale image of a handwritten digit (0-9). Each digit is sometimes store as a long vector of 784 (=28*28) features. The dataset is often used for image classification tasks, where the goal is to train a model to recognize and classify the digits based on the pixel values in the images.

One of many ways to load the MNIST dataset is through the `sklearn` library in Python.



```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784", version=1, as_frame=False)
```

```python
mnist['data']
#target lables 
mnist['target']
mnist['data'][0].shape
#image that was vectorized into 70000

mnist['data'][0].reshape(28, 28)
#display 


import matplotlib.pyplot as plt
#gives a color map to the image that matches with pixel 

plt.imshow(mnist['data'][0].reshape(28, 28), cmap='gray')
mnist['target'][0]

plt.imshow(mnist['data'][1].reshape(28, 28), cmap='gray')
mnist['target'][1]

#image is vectorized into an array 
#numpy array that cant be displayed as easily 

```
